```python
import numpy as np
import pandas as pd
```

```python
dset=pd.read_csv("/content/drive/MyDrive/breast-cancer_csv (1).csv")
dset
```

```python
a=dset.drop(["menopause","breast-quad"],axis=1)
a
```

```python
b=a.drop(["node-caps"],axis=1)
```

```python
b["irradiat"]=[1 if each=="no" else 2 for each in b.irradiat]
b
```

```python
b["breast"]=[0 if each=="right" else 1 for each in b.breast]
b
```

```python
b["Class"]=[0 if each=="no-recurrence-events" else 1 for each in b.Class]
b
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
b["age"]=label_encoder.fit_transform(b["age"])
b["tumor-size"]=label_encoder.fit_transform(b["tumor-size"])
b["inv-nodes"]=label_encoder.fit_transform(b["inv-nodes"])
```

```python
b
```

```python
y=b["irradiat"]
x=b.drop(["irradiat"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(x_train,y_train)
```

```python
rf_pred=dtree.predict(x_test)
rf_pred
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100)
rf.fit(x_train,y_train)
```

```python
rf_pred=rf.predict(x_test)
rf_pred
```
